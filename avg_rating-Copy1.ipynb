```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('df_cleaned.csv') 
genre = pd.read_csv('genres.csv') 
print(df.shape)
display(df.head())
print(genre.shape)
genre.head()
```

```python
rating_genre = pd.concat([df['rating'], genre], axis=1) 
rating_genre
```

```python
def calculate_mean(col1, df=rating_genre, col2='rating'):
    filtered_df = df[df[col1] == 1]
    result = filtered_df.groupby(col1)[col2].mean()
    return result
```

```python
import pandas as pd

# Assuming you have defined the necessary functions and data structures
# col_list is the list of columns you want to process
# rating_genre is the DataFrame containing the ratings and genres

result_df = pd.DataFrame()

for col in rating_genre.columns:
    col_result = calculate_mean(col, rating_genre)
    col_result = col_result.reset_index()  # Reset the index of the col_result
    result_df = pd.concat([result_df, col_result.set_index('genre')], axis=1)

result_df
```

```python
result_df = pd.DataFrame(columns=[])  # Create an empty DataFrame

for col in rating_genre.columns:
    if col != 'rating':
        col_result = calculate_mean(col, rating_genre)
        col_result = col_result.reset_index()  # Reset the index of the col_result
        col_result.columns = [col]  # Rename columns for concatenation
        result_df = pd.concat([result_df, col_result.set_index()], axis=1)

result_df
```

```python
result_df = result_df.drop(['index', 'rating'], axis=1)
result_df
```

```python
result_df = result_df.T
result_df
```

```python
result_df = result_df.reset_index()
result_df
```

```python
result_df.rename(columns={1: 'avg_rating'}, inplace=True)
result_df.rename(columns={'index': 'genre'}, inplace=True)
result_df
```

```python
result_df = result_df.sort_values(by=['avg_rating'], ascending=False)
result_df 
```

```python
sns.barplot(x='genre', y='avg_rating', data=result_df)
plt.figure(figsize=(30, 30))
plt.xticks(rotation=90)
plt.show()
```

### Ratings By Stars

```python
df['ratingsByStars'] = [e.replace("'", "") for e in df['ratingsByStars']]
df['ratingsByStars'] = [e.replace("[", "") for e in df['ratingsByStars']]
df['ratingsByStars'] = [e.replace("]", "") for e in df['ratingsByStars']]

df.head()
```

```python
df[['5*', '4*', '3*', '2*', '1*']] = df['ratingsByStars'].str.split(',', expand=True)

df = df.drop(columns=['ratingsByStars'])

df.head()
```

```python
df.isna().sum()
```

We suppose that NaN values for star ratings are because people only gave a 5 stars rating so we replace them by 0.

```python
df[['5*', '4*', '3*', '2*', '1*']] = df[['5*', '4*', '3*', '2*', '1*']].fillna(0)
df.isna().sum()
```

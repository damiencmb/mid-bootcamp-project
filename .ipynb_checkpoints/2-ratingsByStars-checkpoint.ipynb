```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option('display.max_columns', None)
```

### We have already cleaned all the NaN values execpt the 'likedPercent' column

```python
df = pd.read_csv('df_cleaned_NaN.csv') 
df
```

This column correspond to the the sum of 3 stars or more divided by the sum of the total total stars time 100. We are gonna fill the NaN values with this number, but first we need to seperate the column ratingsByStars in 5, one for each number of stars ratings.

## Ratings by stars

```python
# we first remove the apostrophes and square brackets
df['ratingsByStars'] = [e.replace("'", "") for e in df['ratingsByStars']]
df['ratingsByStars'] = [e.replace("[", "") for e in df['ratingsByStars']]
df['ratingsByStars'] = [e.replace("]", "") for e in df['ratingsByStars']]
df.head()
```

```python
# then we split the values on the coma
df[['5*', '4*', '3*', '2*', '1*']] = df['ratingsByStars'].str.split(',', expand=True)

df = df.drop(columns=['ratingsByStars'])

df
```

```python
df.isna().sum()
```

We suppose that NaN values for star ratings are because people only gave a 5 stars rating so we replace them by 0.

```python
df[['5*', '4*', '3*', '2*', '1*']] = df[['5*', '4*', '3*', '2*', '1*']].fillna(0)
df.isna().sum()
```

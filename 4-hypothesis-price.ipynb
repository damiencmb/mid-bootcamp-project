# We want to know if a book has a better rating if it's cheaper

```python
df = pd.read_csv('df_for_analysis.csv') 
df
```

For a fair comparaison we take only the books rated more than a hundred thousand time.

```python
df = df[(df['numRatings']> 100000)]
```

```python
df
```

```python
df['price'].mean()
```

We will study a sample of the 25% top rated books.

```python
top25 = df['rating'].quantile(q=0.75)
top25
```

```python
top25 = df[(df['rating']> 4.21)]
top25
```

```python
from scipy.stats import ttest_1samp

# H0 = price top rated books <= mean price
# H1 = price top rated books > mean price

stat, pval = ttest_1samp(top25['price'], 6.63)

print('stat is  ', stat)
print('pvalue for the two-tailed test is ', pval/2)
```

The p value is very low but the stat is positive whereas the null hypothesis was 'less than' so we can reject the null hypothesis.

We can also verify it with the average price of our top 25 rated books sample.

```python
top25['price'].mean()
```

# A book is not better rated because it has a low price.

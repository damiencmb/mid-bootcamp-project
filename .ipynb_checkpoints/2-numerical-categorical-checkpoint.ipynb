```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('df_cleaned.csv') 
df
```

# Categorical columns

```python
categorical = df.select_dtypes(include=object)
```

```python
for col in categorical.columns:
    display(categorical[col].value_counts())
```

```python
categorical.columns.tolist()
```

We will not focus on 'bookId', 'title', and 'publisher' columns because it as to many unique values which we can not group.\
The top unique value in the 'publisher' column appaears only 572, which is nothin compare of the total of the dataset, we let this column as it is, we will not need it for the analysis.
We keep 'author' and 'genres' columns for later.\
The column 'publishDate is supposed to be a datetime format so we update it.

```python
categorical = df[['series', 'language', 'bookFormat', 'severalPublications']]
```

```python
for col in categorical.columns:
    display(categorical[col].value_counts())
```

We have already formatted the column 'series' when we cleaned so we keep it like this.

### 1. Language

Most of the books are in English, all the others languages will be grouped in 'other'

```python
categorical['language'] = np.where(categorical['language'].isin(['English']),'English','Other')
```

```python
categorical['language'].value_counts()
```

### 2. Book format

Most of the book are paperback or hardcover, all the others formats will be grouped in 'other'

```python
categorical['bookFormat'] = np.where(categorical['bookFormat'].isin(['Paperback', 'Hardcover']),categorical['bookFormat'],'Other')
```

```python
categorical['bookFormat'].value_counts()
```

```python
def cntplot_cat(df):
    for column in df.columns:
        sns.countplot(x=df[column])
        plt.xticks(rotation=90)
        plt.show()
    return None
   

cntplot_cat(categorical)
```

 ### 3. Authors

In the 'author' column we can see that some have also other names, like illustartor.\
We are gonna keep only the first name.

```python
# we keep only the 2 first names in the colum and we drop the coma
df['author'] = df['author'].str.split().str[:2].str.join(' ').str.replace(',','')
```

```python
df.head()
```

### 4. Publish date

```python
df['publishDate'] = pd.to_datetime(df['publishDate'], errors='coerce')
```

```python
df.dtypes
```

# Numerical columns

```python
numerical = df.select_dtypes(include=np.number)
numerical
```

## Outliers

```python
for col in numerical:
    numerical[[col]].boxplot()
    plt.show()
```

Only the column 'pages' seems to have problematic outliers, let's remove them 

```python
iqr = np.percentile(numerical['pages'],75) - np.percentile(numerical['pages'],25)
upper_limit = np.percentile(numerical['pages'],75) + 5*iqr
lower_limit = np.percentile(numerical['pages'],25) - 5*iqr
df_no_outliers = numerical[(numerical['pages']>lower_limit) & (numerical['pages']<upper_limit)].copy()
sns.displot(df_no_outliers['pages'])
plt.show()
```

```python
numerical['pages'] = df_no_outliers['pages']
```

```python
numerical[['pages']].boxplot()
```

```python
numerical.isna().sum()
```

Now we fill the 'pages' column NaN values with the mean.

```python
pagesmean = numerical['pages'].mean()
```

```python
numerical['pages'] = numerical['pages'].fillna(pagesmean)
```

```python
numerical.isna().sum()
```

```python
numerical[(numerical['pages']==0)]
```

We have also noticed that some books have 0 pages, which mmight be an error, so we replace them by the mean.

```python
pagesmean = numerical['pages'].mean()
pagesmean
```

```python
numerical['pages']=np.where(numerical['pages'].isin([0]),pagesmean,numerical['pages'])
```

```python
numerical[(numerical['pages']==0)]
```

### Now we have removed the outliers, let's check for correlations

```python
correlations_matrix = numerical.corr()
sns.heatmap(correlations_matrix, annot=True)
```

We can see that ratins are corralated with the percentage of liked books which can be explained by the fact that liked books are those with more than 2 stars.


```python
for column in numerical:
    sns.distplot(numerical[column])
    plt.show()
```

We can see a bump on 0, probably because doens't have any rating so they are rated 0.\
The pic at 5 may due to some books have only a 5 stars rating.

The shape of likedBooks looks a bit like ratings, as we see the correlation between each others.

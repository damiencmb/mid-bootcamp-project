```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('df_cleaned.csv') 
genre = pd.read_csv('genres.csv') 
print(df.shape)
display(df.head())
print(genre.shape)
genre.head()
```

```python
rating_genre = pd.concat([df['rating'], genre], axis=1) 
rating_genre
```

```python
fiction = rating_genre[(rating_genre['Fiction']==1)] 
fiction = fiction[['rating','Fiction']]
fiction
# fiction['Fiction'].value_counts()
```

```python
sns.countplot(x=fiction['rating'])
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(fiction['rating'], bins=50, kde=True) # bins = intervalle données x , kde=True : pour afficher une ligne
plt.show()
```

```python
fiction.groupby(['Fiction']).agg({'rating':np.mean})
```

```python
def calculate_mean(col1, df=rating_genre, col2='rating'):
    filtered_df = df[df[col1] == 1]
    result = filtered_df.groupby(col1)[col2].mean()
    return result

calculate_mean('Romance')
```

```python
for col in rating_genre.columns:
    if col != 'rating':  
        result = pd.DataFrame(calculate_mean(col))
        display(result)
 
   
```

```python
result_df = pd.DataFrame(columns=['index', 'rating'])  # Create an empty DataFrame

for col in rating_genre.columns:
    if col != 'rating':
        col_result = calculate_mean(col, rating_genre)
        col_result = col_result.reset_index()  # Reset the index of the col_result
        col_result.columns = ['index', col]  # Rename columns for concatenation
        result_df = pd.concat([result_df, col_result.set_index('index')], axis=1)

result_df
```

```python
result_df = result_df.drop(['index', 'rating'], axis=1)
result_df = result_df.T
```

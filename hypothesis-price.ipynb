## We want to know if a low price can make a book well rated


We take a sample of the best rated books as the 25% best rated books.

```python
top_rated = df[df['rating'] > df['rating'].quantile(0.75)]
top_rated             
```

```python
from scipy.stats import ttest_1samp

# H0 = mean price top rated books <= mean price
# H1 = mean price top rated books > mean price

stat, pval = ttest_1samp(top_rated['price'], 9.66)

print('stat is  ', stat)
print('pvalue for the two-tailed test is ', pval/2)
```

The p value is very low and the stat is positiv whereas the null hypothesis was less than so we can reject the Null hypothesis.

```python
mean_price = df['price'].mean()
print("The mean price of all books is :", mean_price)
mean_price_top_rated = top_rated['price'].mean()
print("The mean price the best rated books is :", mean_price_top_rated)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('df_cleaned.csv') 
print(df.shape)
df.head()
```

```python
df['genres']=df['genres'].str.replace('\[','')
df['genres']=df['genres'].str.replace('\]','')
df['genres']=df['genres'].str.replace('\'','')
```

```python
genres = df['genres'].str.get_dummies(sep=',')
```

```python
sumgenre = genres.sum().sort_values(ascending=False)
```

```python
top = sumgenre.head(1)
```

```python
top
```

```python
hypo = pd.concat([df['rating'], genres['Fiction']], axis=1) 
```

```python
hypo
```

```python
from scipy.stats import ttest_1samp

# H0 = Fiction = 1
# H1 = Fiction < 1

stat, pval = ttest_1samp(hypo, 1)

print('stat is  ', stat)
print('pvalue for the two-tailed test is ', pval)
```

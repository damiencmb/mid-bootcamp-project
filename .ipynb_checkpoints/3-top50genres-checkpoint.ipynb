## Let's have a look on genres distibution

```python
df = pd.read_csv('df_for_analysis.csv') 
print(df.shape)
df.head()
```

We first need to clean the column by removing the square brackets and quotation mark

```python
df['genres']=df['genres'].str.replace('\[','')
df['genres']=df['genres'].str.replace('\]','')
df['genres']=df['genres'].str.replace('\'','')
print(df.shape)
df.head()
```

There are several genre for a same book so we need to get all individual genre.\
For this we enocde all the genre by seprating it on the coma.

```python
genres = df['genres'].str.get_dummies(sep=',')
```

```python
rating_genre = pd.concat([df[['rating','numRatings']], topgenre_encoded], axis=1)
rating_genre                           
```

For a fairer comparaison, we only take books that have more than 100.000 ratings.

```python
rating_genre = rating_genre[(rating_genre['numRatings']>100000)]
```

```python
rating_genre
```

```python
# we create a function to calcul the mean of a genre 
def calculate_mean(col1, df=rating_genre, col2='rating'):
    filtered_df = df[df[col1] == 1]
    mean = filtered_df.groupby(col1)[col2].mean()
    return mean
```

```python
# we loop through all the genre and apply the function

# we create an empty DataFrame
mean_genre = pd.DataFrame(columns=['index', 'rating'])

for col in rating_genre.columns:
    if col != 'rating':
        col_result = calculate_mean(col, rating_genre)
        col_result = col_result.reset_index()  # we reset the index of the col_result
        col_result.columns = ['index', col]  # we rename columns for concatenation
        mean_genre = pd.concat([mean_genre, col_result.set_index('index')], axis=1)

mean_genre
```

```python
mean_genre = mean_genre.drop(['index', 'rating','numRatings'], axis=1)
mean_genre
```

```python
mean_genre= mean_genre.T
mean_genre
```

```python
mean_genre = mean_genre.reset_index()
mean_genre
```

```python
mean_genre.rename(columns={1: 'avg_rating'}, inplace=True)
mean_genre.rename(columns={'index': 'genre'}, inplace=True)
mean_genre
```

```python
mean_genre = mean_genre.sort_values(by=['avg_rating'], ascending=False)
mean_genre
```

```python
plt.figure(figsize=(30, 10))
sns.barplot(x='genre', y='avg_rating', data=mean_genre)
plt.xticks(rotation=90,fontsize=25)
plt.show()
```

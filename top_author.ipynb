```python
df = pd.read_csv('df_cleaned.csv') 
print(df.shape)
df.head()
```

We notice that the author who got the best rating only have a few number rating so it's not very representative that's why we'll keep only the top 25% number of ratings.

```python
most_rated = df[df['numRatings'] > df['numRatings'].quantile(0.75)]
most_rated
```

```python
most_rated['numRatings'].min()
```

```python
most_rated['numRatings'].max()
```

```python
topmostrated = most_rated.sort_values(by='rating', ascending=False).head(50)
topmostrated 
```

```python
avgtopmostrated = topmostrated.groupby(['author']).agg({'rating':'mean'})
avgtopmostrated
```

```python
avgtopmostrated = avgtopmostrated.sort_values(by='rating', ascending=False).reset_index()
avgtopmostrated
```

```python
plt.figure(figsize=(30, 10))
sns.barplot(x='author', y='rating', data=avgtopmostrated)
plt.xticks(rotation=90,fontsize=25)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('df_cleaned.csv') 
print(df.shape)
df.head()
```

```python
df['genres']=df['genres'].str.replace('\[','')
df['genres']=df['genres'].str.replace('\]','')
df['genres']=df['genres'].str.replace('\'','')
print(df.shape)
df.head()
```

```python
genres = df['genres'].str.get_dummies(sep=',')
```

```python
genres.to_csv('genres_encoded.csv', index=False)
```

```python
sumgenre = genres.sum().sort_values(ascending=False)
topgenre = sumgenre.head(50)
topgenre
```

```python
topgenre = topgenre.index.tolist()
topgenre
```

```python
topgenres = genres[[' Fiction',
 ' Romance',
 ' Contemporary',
 ' Fantasy',
 ' Young Adult',
 ' Adult',
 ' Novels',
 ' Audiobook',
 ' Adventure',
 ' Historical',
 'Fiction',
 ' Literature',
 ' Classics',
 ' Mystery',
 ' Nonfiction',
 'Fantasy',
 ' Historical Fiction',
 ' Paranormal',
 ' Childrens',
 ' Magic',
 ' Thriller',
 ' Humor',
 ' Science Fiction',
 ' Contemporary Romance',
 ' Crime',
 ' Suspense',
 'Romance',
 ' Science Fiction Fantasy',
 ' Supernatural',
 ' Middle Grade',
 'Young Adult',
 ' Mystery Thriller',
 ' Chick Lit',
 ' Urban Fantasy',
 ' Teen',
 ' Adult Fiction',
 ' Biography',
 ' History',
 ' Paranormal Romance',
 'Nonfiction',
 ' Literary Fiction',
 ' British Literature',
 ' Drama',
 ' Realistic Fiction',
 'Historical Fiction',
 ' Philosophy',
 ' Short Stories',
 ' Memoir',
 ' Religion',
 ' Horror',
 ' New Adult',
 ' 20th Century',
 ' War',
 'Mystery',
 ' Erotica',
 ' American',
 ' Juvenile',
 ' Family',
 ' School',
 ' Biography Memoir']]
```

We notice that there are several times the same genre name so we group the columns together.

```python
topgenres[' Fiction'] = topgenres[' Fiction']+ topgenres['Fiction']
topgenres = topgenres.drop(columns=['Fiction'])
```

```python
topgenres[' Romance'] = topgenres[' Romance']+ topgenres['Romance']
topgenres = topgenres.drop(columns=['Romance'])
```

```python
topgenres[' Fantasy'] = topgenres[' Fantasy']+ topgenres['Fantasy']
topgenres = topgenres.drop(columns=['Fantasy'])
```

```python
topgenres[' Young Adult'] = topgenres[' Young Adult']+ topgenres['Young Adult']
topgenres = topgenres.drop(columns=['Young Adult'])
```

```python
topgenres[' Historical Fiction'] = topgenres[' Historical Fiction']+ topgenres['Historical Fiction']
topgenres = topgenres.drop(columns=['Historical Fiction'])
```

```python
topgenres[' Mystery'] = topgenres[' Mystery']+ topgenres['Mystery']
topgenres = topgenres.drop(columns=['Mystery'])
```

```python
topgenres[' Nonfiction'] = topgenres[' Nonfiction']+ genres['Nonfiction']
topgenres = topgenres.drop(columns=['Nonfiction'])
```

We can also conbine 'Biography' and ' Biography Memoir'.

```python
topgenres[' Biography' ] = topgenres[' Biography']+ topgenres[' Biography Memoir']
topgenres = topgenres.drop(columns=[' Biography Memoir'])
```

Same for 'Historical' and 'Historical fiction'

```python
topgenres[' Historical Fiction' ] = topgenres[' Historical Fiction']+ topgenres[' Historical']
topgenres = topgenres.drop(columns=[' Historical'])
```

```python
sumtopgenre = topgenres.sum().sort_values(ascending=False)
topgenre = sumtopgenre.head(50)
topgenre
```

```python
dftopgenre = pd.concat([df, genres], axis=1) 
dftopgenre = dftopgenre.drop(columns=['genres'])
dftopgenre 
```
